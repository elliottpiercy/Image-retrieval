```python
import numpy as np
from os import listdir
from os.path import isfile, join
import cv2

def get_images(seed = 1):
    path = 'G:/Elliott/train_sm/'
    images = np.array([])
    count = 0

    for f in listdir(path):
        if isfile(join(path, f)):
            data = cv2.cvtColor(cv2.imread(path+f), cv2.COLOR_BGR2GRAY)
            data = cv2.resize(data, dsize=(150,150), interpolation=cv2.INTER_CUBIC)
            images = np.append(images,np.ravel(data))
            count +=1

    images = np.reshape(images, (count,150*150))
    np.random.seed(seed)
    np.random.shuffle(images)
    return images



class data():
    
    def __init__(self,data, train, validation, test):
        
        if train + validation + test != 1:
            raise ValueError("Data split must sum to 1.")
            
        self.data = data
        self.train = self.train(data[:int(len(data)*train)],np.arange(int(len(data)*train)))
        
        self.validation = self.validation(data[int(len(data)*train):int(len(data)*train)+int(len(data)*validation)],
                                          np.arange(int(len(data)*validation)))
        
        self.test = self.test(data[int(len(data)*train)+int(len(data)*validation):],
                              np.arange(int(len(data)*test)))
        
    
    # Class for training set variables batch method
    class train():
    
        def __init__(self,data,labels):
            self.data = data
            self.labels = labels
            self.num_examples = data.shape[0]
            self.batch_count = 0


        def next_batch(self,batch_size):
            next_batch = self.data[self.batch_count:self.batch_count + batch_size]
            self.batch_count += batch_size
            return next_batch   
        
        
    # Class for validation set variables
    class validation():
    
        def __init__(self,data,labels):
            self.data = data
            self.labels = labels
            self.num_examples = data.shape[0]

        
    # Class for test set variables 
    class test():
            
        def __init__(self,data,labels):
            self.data = data
            self.labels = labels
            self.num_examples = data.shape[0]
        
        
        
        
        
imgs = get_images()
satellite = data(imgs, train = 0.7, validation = 0.15, test = 0.15)
```
